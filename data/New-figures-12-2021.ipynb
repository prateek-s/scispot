```python
import json
import pandas as pd
import numpy as np
import time
import datetime
import dateutil.parser 
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib notebook
%matplotlib inline
import math
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from scipy.integrate import quad
from scipy.optimize import curve_fit
import glob
import inspect
import os
import seaborn as sns
```

```python
#linestyles = ['-', '--', '-', '--', '-', '--', '-', '--']
linestyles = ['-',':', '--', '-.']

#markerstyles = ["+", "x", "o", "v", "^", "<", ">", "1"]
markerstyles = ["o", "s", "^", "p", "v", "d", "*", "<"]
colors = ['blue', 'orange', 'green', 'red']
colors2 = [ 'black','cyan', 'magenta','brown']



fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))

plot_labels = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

palette = np.array(sns.color_palette("hls", len(plot_labels)))

count=0
for label in plot_labels:
    data_gt = np.loadtxt(os.path.join(os.getcwd(), 'runtime-data-by-group/'+label+'.txt'))
    ax.plot(data_gt[:,0],data_gt[:,1], label=label, marker=markerstyles[count],markerfacecolor='none', 
            linestyle='none', color=palette[count])
    data = np.loadtxt(os.path.join(os.getcwd(), 'runtime-data-by-group/blended_exp_model/'+label+'.txt'))
    ax.plot(data[:,0],data[:,1], label='_Hidden', color=palette[count])

    count += 1
    
    
ax.legend(ncol=2)
ax.set_xlabel('Time to Preemption (Hours)', fontsize=14)
ax.set_ylabel('CDF', fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=11)
ax.tick_params(axis='both', which='minor', labelsize=11)
```
